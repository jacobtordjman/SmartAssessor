### Imports, Load Model & Define Solver

```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
from peft import PeftModel
import json
import pdfplumber

# 1. Configuration
BASE_MODEL   = "WizardLMTeam/WizardMath-7B-V1.1"
ADAPTER_PATH = "/content/drive/MyDrive/wizardmath-lora-checkpoints/checkpoint-1800"  # where your LoRA adapters live
DEVICE       = "cuda" if torch.cuda.is_available() else "cpu"

# 2. Load tokenizer
print("Loading tokenizer")
tokenizer = AutoTokenizer.from_pretrained(BASE_MODEL)

# 3. Load base model in 8-bit with CPU offload
print("Loading base model with 8-bit + CPU offload")
bnb_config = BitsAndBytesConfig(
    load_in_8bit=True,
    llm_int8_enable_fp32_cpu_offload=True,
)
base = AutoModelForCausalLM.from_pretrained(
    BASE_MODEL,
    quantization_config=bnb_config,
    device_map="auto",
)
# 4. Attach your fine-tuned LoRA adapters
print("Attaching LoRA adapters")
model = PeftModel.from_pretrained(base, ADAPTER_PATH)
model.to(DEVICE)
model.eval() # Set model to evaluation mode
print(f" Model loaded onto {DEVICE}")

# 5. Define the solver function
SYSTEM_PROMPT = """You are an expert in linear algebra. Provide a clear, step-by-step derivation.
### Problem:
{question}
### Solution:"""

def solve_problem(question: str,
                  max_new_tokens: int = 256,
                  temperature: float = 0.0) -> str:
    prompt = SYSTEM_PROMPT.format(question=question)
    inputs = tokenizer(prompt, return_tensors="pt").to(DEVICE)
    outputs = model.generate(
        **inputs,
        max_new_tokens=max_new_tokens,
        temperature=temperature,
        eos_token_id=tokenizer.eos_token_id,
        pad_token_id=tokenizer.eos_token_id,
    )
    text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    # strip off the prompt itself
    return text[len(prompt):].strip()
```

## Comparator Tool

```python
#  Cell: Improved Comparator 

import re
import sympy as sp

def extract_solution_block(text: str) -> str:
    m = re.search(r"Solution:\s*(.+)", text, flags=re.IGNORECASE | re.DOTALL)
    if not m:
        raise ValueError("No 'Solution:' section found.")
    return m.group(1).strip()

def extract_final_value(solution: str) -> str:
    """
    Heuristic: find the first number/rational in the last line.
    E.g.
      "The answer is: -8."   -> "-8"
      "... = 5/2"            -> "5/2"
      "... eigenvalues 3, 2" -> "3, 2"  (handles comma-separated)
    """
    # Take the last non-empty line
    lines = [ln.strip() for ln in solution.splitlines() if ln.strip()]
    last = lines[-1]

    # Regex to match integers, decimals, rationals, or comma-separated lists
    # e.g. -8, 2.5, 5/2, 3,2
    pattern = r"-?\d+(?:\.\d+)?(?:/\d+(?:\.\d+)?)?(?:\s*,\s*-?\d+(?:\.\d+)?(?:/\d+(?:\.\d+)?)?)*"
    m = re.search(pattern, last)
    if m:
        return m.group(0)
    # fallback: return the whole line
    return last

def compare_final(student_text: str, model_text: str) -> bool:
    stud_block = extract_solution_block(student_text)
    mod_block  = model_text  # solver output is just the solution text

    stud_val = extract_final_value(stud_block)
    mod_val  = extract_final_value(mod_block)

    print(f"Student final value: {stud_val}")
    print(f"Model   final value: {mod_val}")

    # Compare as SymPy expressions if possible
    try:
        s = sp.sympify(stud_val)
        m = sp.sympify(mod_val)
        return bool(sp.simplify(s - m) == 0)
    except Exception as e:
        print("Sympy parse error:", e)
        # For comma-separated lists, compare sets of rationals
        if "," in stud_val and "," in mod_val:
            s_list = [sp.sympify(x) for x in stud_val.split(",")]
            m_list = [sp.sympify(x) for x in mod_val.split(",")]
            return set(s_list) == set(m_list)
        # fallback to lowercase stripped compare
        return stud_val.strip().lower() == mod_val.strip().lower()
```

```python
import json
import re
from SmartAssessor.backend.app.solver import solve_problem
# reuse compare_final from your comparator cell

def grade(prompt: str) -> str:
    """
    Agent tool that:
      - Expects a single string containing:
         'QUESTION:' followed by the problem text,
         then 'STUDENT SOLUTION:' followed by the student's answer.
      - Runs solver + comparator under the hood.
      - Returns a JSON string with fields:
         question, student, model, correct, feedback
    """
    # Parse the sections
    q_match = re.search(r"QUESTION:(.+?)STUDENT SOLUTION:", prompt, flags=re.DOTALL)
    s_match = re.search(r"STUDENT SOLUTION:(.+)$", prompt, flags=re.DOTALL)
    if not q_match or not s_match:
        return json.dumps({"error":"Prompt must include QUESTION:... STUDENT SOLUTION:..."})


    question_text = q_match.group(1).strip()
    student_text  = s_match.group(1).strip()

    # Solve
    canonical = solve_problem(question_text)
    # Compare
    comp = compare_final(student_text, canonical)

    # Build feedback
    feedback = " Correct!" if comp["correct"] else (
        f" Incorrect. Your answer {comp['student']} vs correct {comp['model']}."
    )

    result = {
        "question": question_text,
        "student": comp["student"],
        "model": comp["model"],
        "correct": comp["correct"],
        "feedback": feedback,
        "canonical_solution": canonical
    }
    return json.dumps(result)
```

```python
# Core
import json
import re
import math
from pathlib import Path
from typing import Any, Dict, List

# PDF
import pdfplumber  # pip install pdfplumber

# HF runtime (works with PEFT models)
from transformers import pipeline

# 
# Single-Agent with Tool Use (no reload; uses preloaded `model` + `tokenizer`)
# Tools: extract_from_pdf, solve, grade, get_qa, append_result, save_results,
#        evaluate_pdf (one-shot), evaluate_inline (one-shot single pair)
# 

# Preconditions: `model` (PEFT LoRA attached) and `tokenizer` are already in GPU RAM.

# Build a generation pipeline that supports PEFT models
try:
    _ = model  # noqa
    _ = tokenizer  # noqa
except NameError as e:
    raise RuntimeError("Expected `model` and `tokenizer` to be preloaded with LoRA attached.") from e

if getattr(tokenizer, "pad_token_id", None) is None and getattr(tokenizer, "eos_token_id", None) is not None:
    tokenizer.pad_token = tokenizer.eos_token

# Use the tasked pipeline  this handles PeftModelForCausalLM correctly
_llm = pipeline(
    task="text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=512,
    temperature=0.0,
    do_sample=False,
    return_full_text=False,
    pad_token_id=tokenizer.eos_token_id,
    eos_token_id=tokenizer.eos_token_id,
)

#  JSON enforcement constants & few-shot examples 
JSON_TAG_OPEN = "<json>"
JSON_TAG_CLOSE = "</json>"

FEW_SHOT = (
    "Examples of valid responses (exactly one per step):\n"
    f"{JSON_TAG_OPEN}" + '{"tool_call":{"name":"extract_from_pdf","args":{"pdf_path":"/content/drive/MyDrive/linear_questions_student_ans.pdf"}}}' + f"{JSON_TAG_CLOSE}\n"
    f"{JSON_TAG_OPEN}" + '{"tool_call":{"name":"get_qa","args":{"index":0}}}' + f"{JSON_TAG_CLOSE}\n"
    f"{JSON_TAG_OPEN}" + '{"tool_call":{"name":"solve","args":{"problem":"Compute det([[1,2],[3,4]])"}}}' + f"{JSON_TAG_CLOSE}\n"
    f"{JSON_TAG_OPEN}" + '{"tool_call":{"name":"grade","args":{"problem":"Compute det([[1,2],[3,4]])","student_answer":"-2","expected_answer":"-2"}}}' + f"{JSON_TAG_CLOSE}\n"
    # single-pair one-shot example so the agent does what your smoke test expects
    f"{JSON_TAG_OPEN}" + '{"tool_call":{"name":"evaluate_inline","args":{"problem":"Compute det([[1,2],[3,4]])","student_answer":"-2"}}}' + f"{JSON_TAG_CLOSE}\n"
    f"{JSON_TAG_OPEN}" + '{"final":{"summary":"Evaluated all items and saved results to evaluator_results.json","next_actions":"Open the JSON file to review scores."}}' + f"{JSON_TAG_CLOSE}"
)

#  Helpers 

def _complete(prompt: str, max_new_tokens: int = 512, **gen_kwargs) -> str:
    out = _llm(prompt, max_new_tokens=max_new_tokens, **gen_kwargs)
    return out[0]["generated_text"]

def _extract_json(text: str) -> Dict[str, Any]:
    # Prefer <json>...</json> blocks
    if JSON_TAG_OPEN in text and JSON_TAG_CLOSE in text:
        inner = text.split(JSON_TAG_OPEN, 1)[1].split(JSON_TAG_CLOSE, 1)[0].strip()
        return json.loads(inner)
    # Fallback: outermost braces
    start = text.find("{"); end = text.rfind("}")
    if start != -1 and end != -1 and end > start:
        return json.loads(text[start:end+1])
    raise ValueError("No JSON object found in model output.")

def _coerce_to_json(raw: str, tools_text: str) -> Dict[str, Any]:
    """Ask the model to convert arbitrary reply into a valid JSON shape."""
    repair_prompt = (
        "Convert the following assistant reply into ONE valid response JSON. "
        "Valid shapes only:\n"
        '1) {"tool_call":{"name":<tool_name_from_catalog>,"args":{...}}}\n'
        '2) {"final":{"summary":"...", "next_actions":"..."}}\n\n'
        "Use only tools from this catalog and respect their arg schemas:\n"
        f"{tools_text}\n\n"
        f"Output ONLY inside {JSON_TAG_OPEN}{JSON_TAG_CLOSE}, no extra text.\n\n"
        "ASSISTANT_REPLY_TO_FIX:\n"
        f"{raw}"
    )
    fixed = _complete(repair_prompt, max_new_tokens=384)
    return _extract_json(fixed)

def _minified_json(d: Dict[str, Any]) -> str:
    return json.dumps(d, separators=(",", ":"), ensure_ascii=False)

#  State (blackboard) 
STATE: Dict[str, Any] = {
    "qas": [],       # List[{"page_index": int, "problem": str, "student_answer": str}]
    "results": [],   # List[{"index": int, "expected_answer": str, "verdict": str, ...}]
}

#  Tools 

def tool_extract_from_pdf(pdf_path: str) -> Dict[str, Any]:
    qas: List[Dict[str, Any]] = []
    with pdfplumber.open(pdf_path) as pdf:
        for i, page in enumerate(pdf.pages):
            text = (page.extract_text() or "").strip()
            if not text:
                continue
            problem, student = None, None
            if "QUESTION:" in text and "STUDENT ANSWER:" in text:
                chunk = text.split("QUESTION:", 1)[1]
                problem, student = chunk.split("STUDENT ANSWER:", 1)
            else:
                parts = [blk.strip() for blk in re.split(r"\n\s*\n", text) if blk.strip()]
                if parts:
                    problem = parts[0]
                    student = "\n\n".join(parts[1:]) if len(parts) > 1 else ""
            if problem:
                qas.append({
                    "page_index": i,
                    "problem": problem.strip(),
                    "student_answer": (student or "").strip()
                })
    STATE["qas"] = qas
    return {"status": "ok", "num_items": len(qas)}

def tool_get_qa(index: int) -> Dict[str, Any]:
    qas = STATE.get("qas", [])
    if not (0 <= index < len(qas)):
        return {"error": f"index {index} out of range (0..{len(qas)-1})"}
    return qas[index]

def _llm_json_instruct(prompt_instructions: str, fields_desc: str) -> Dict[str, Any]:
    """Ask for strict JSON wrapped in tags; attempt repair if needed."""
    prompt = (
        "You are a precise math assistant. "
        f"ALWAYS answer with a single JSON object wrapped by {JSON_TAG_OPEN} and {JSON_TAG_CLOSE}; no extra text.\n\n"
        f"{prompt_instructions}\n\n"
        f"Return JSON with fields: {fields_desc}\n"
        "Respond with minified JSON (no spaces/newlines)."
    )
    out = _complete(prompt, max_new_tokens=512)
    try:
        return _extract_json(out)
    except Exception:
        # Try a quick self-repair pass
        fixed = _complete(
            prompt + "\nIf your last answer was not valid JSON, convert it now to valid JSON inside the tags.",
            max_new_tokens=384
        )
        return _extract_json(fixed)

def tool_solve(problem: str) -> Dict[str, Any]:
    instr = (
        "Solve the university-level linear algebra problem. "
        "Give the final answer clearly (exact if possible), plus a brief 12 sentence justification."
        f"\n\nPROBLEM:\n{problem}"
    )
    fields = '{"expected_answer":"string","explanation":"string"}'
    return _llm_json_instruct(instr, fields)

def tool_grade(problem: str, student_answer: str, expected_answer: str) -> Dict[str, Any]:
    instr = (
        "Grade the STUDENT_ANSWER for the PROBLEM by comparing it to EXPECTED_ANSWER. "
        "Allow equivalent forms; keep feedback concise (<=3 sentences)."
        f"\n\nPROBLEM:\n{problem}\n\nEXPECTED_ANSWER:\n{expected_answer}\n\nSTUDENT_ANSWER:\n{student_answer}"
    )
    fields = (
        '{"correctness":"number(0..1)","reasoning_quality":"number(0..1)",'
        '"final_score":"integer(0..100)","verdict":"correct|partially_correct|incorrect",'
        '"feedback":"string"}'
    )
    return _llm_json_instruct(instr, fields)

def tool_append_result(index: int, result: Dict[str, Any]) -> Dict[str, Any]:
    STATE["results"].append({"index": index, **result})
    return {"status": "ok", "results_len": len(STATE["results"])}

def tool_save_results(path: str) -> Dict[str, Any]:
    p = Path(path)
    p.write_text(_minified_json(STATE["results"]), encoding="utf-8")
    return {"status": "ok", "saved_to": str(p.resolve())}

def tool_evaluate_pdf(pdf_path: str, out_json_path: str = "evaluator_results.json") -> Dict[str, Any]:
    tool_extract_from_pdf(pdf_path)
    qas = STATE["qas"]
    for idx, item in enumerate(qas):
        sol = tool_solve(item["problem"])
        grd = tool_grade(item["problem"], item["student_answer"], sol["expected_answer"])
        tool_append_result(idx, {
            "expected_answer": sol["expected_answer"],
            "explanation": sol["explanation"],
            **grd
        })
    save = tool_save_results(out_json_path)
    return {"status": "ok", "num_items": len(qas), **save}

# NEW: single-pair one-shot so the agent can pass your smoke test cleanly
def tool_evaluate_inline(problem: str, student_answer: str) -> Dict[str, Any]:
    """Solve + grade one pair and append to results in a single call."""
    sol = tool_solve(problem)
    grd = tool_grade(problem, student_answer, sol["expected_answer"])
    result = {
        "expected_answer": sol["expected_answer"],
        "explanation": sol["explanation"],
        **grd
    }
    tool_append_result(len(STATE["results"]), result)
    return {"status": "ok", "result": result, "results_len": len(STATE["results"])}

#  Tool Registry 

TOOL_REGISTRY: Dict[str, Any] = {
    "extract_from_pdf": {
        "fn": tool_extract_from_pdf,
        "desc": "Parse a PDF of problems+answers into a list of {problem, student_answer}.",
        "schema": {"type":"object","properties":{"pdf_path":{"type":"string"}},"required":["pdf_path"]}
    },
    "get_qa": {
        "fn": tool_get_qa,
        "desc": "Fetch a single QA item by index from the parsed PDF.",
        "schema": {"type":"object","properties":{"index":{"type":"integer"}},"required":["index"]}
    },
    "solve": {
        "fn": tool_solve,
        "desc": "Solve a given linear algebra problem and return the expected_answer.",
        "schema": {"type":"object","properties":{"problem":{"type":"string"}},"required":["problem"]}
    },
    "grade": {
        "fn": tool_grade,
        "desc": "Grade a student's answer against an expected answer.",
        "schema": {"type":"object","properties":{
            "problem":{"type":"string"},
            "student_answer":{"type":"string"},
            "expected_answer":{"type":"string"}
        },"required":["problem","student_answer","expected_answer"]}
    },
    "append_result": {
        "fn": tool_append_result,
        "desc": "Append a grading result (dict) for a given index to the global results list.",
        "schema": {"type":"object","properties":{
            "index":{"type":"integer"},
            "result":{"type":"object"}
        },"required":["index","result"]}
    },
    "save_results": {
        "fn": tool_save_results,
        "desc": "Save the global results list to a JSON file.",
        "schema": {"type":"object","properties":{"path":{"type":"string"}},"required":["path"]}
    },
    "evaluate_pdf": {
        "fn": tool_evaluate_pdf,
        "desc": "One-shot: extract from PDF, solve+grade all, save to JSON.",
        "schema": {"type":"object","properties":{
            "pdf_path":{"type":"string"},
            "out_json_path":{"type":"string"}
        },"required":["pdf_path"]}
    },
    # NEW registry entry
    "evaluate_inline": {
        "fn": tool_evaluate_inline,
        "desc": "Solve+grade a single PROBLEM/STUDENT_ANSWER pair and append to results.",
        "schema": {"type":"object","properties":{
            "problem":{"type":"string"},
            "student_answer":{"type":"string"}
        },"required":["problem","student_answer"]}
    },
}

def _tool_catalog_text() -> str:
    lines = []
    for name, spec in TOOL_REGISTRY.items():
        lines.append(f"- {name}: {spec['desc']}\n  JSON schema: {json.dumps(spec['schema'], separators=(',',':'))}")
    return "\n".join(lines)

def _is_valid_shape(msg: Dict[str, Any]) -> bool:
    if "tool_call" in msg and isinstance(msg["tool_call"], dict):
        return "name" in msg["tool_call"] and "args" in msg["tool_call"]
    if "final" in msg and isinstance(msg["final"], dict):
        return "summary" in msg["final"] and "next_actions" in msg["final"]
    return False

#  Agent Loop (robust JSON + repair + must-use-tool) 

AGENT_SYSTEM = (
    "You are a single autonomous agent for linear-algebra evaluation. "
    "You can call TOOLS to accomplish the GOAL. At each step, respond with only ONE of the following JSON shapes:\n"
    '1) {"tool_call":{"name":<tool_name>,"args":{...}}}\n'
    '2) {"final":{"summary":"...", "next_actions":"..."}}\n'
    f"Always output your single JSON wrapped between {JSON_TAG_OPEN} and {JSON_TAG_CLOSE} with nothing else outside. "
    "You MUST call at least one tool before returning a final; a final with zero tool calls is invalid. "
    "If the GOAL contains explicit PROBLEM and STUDENT_ANSWER, prefer calling 'evaluate_inline'. "
    "If evaluating a full PDF, prefer calling 'evaluate_pdf'.\n\n"
    f"{FEW_SHOT}"
)

def run_agent(goal: str, max_steps: int = 12) -> Dict[str, Any]:
    history: List[Dict[str, Any]] = []
    tools_text = _tool_catalog_text()
    tools_used = 0  # enforce 1 tool call

    for step in range(max_steps):
        obs_text = _minified_json({"history": history[-6:], "results_len": len(STATE["results"])}) if history else "{}"
        prompt = (
            f"{AGENT_SYSTEM}\n\nGOAL:\n{goal}\n\nTOOLS:\n{tools_text}\n\n"
            f"OBSERVATION_CONTEXT:{obs_text}\n\n"
            f"Respond now with exactly one JSON wrapped by {JSON_TAG_OPEN}...{JSON_TAG_CLOSE}."
        )

        # Try 1: normal greedy
        raw = _complete(prompt, max_new_tokens=512)

        # Parse or repair
        try:
            msg = _extract_json(raw)
        except Exception:
            # Try 2: nudge + beam search
            raw2 = _complete(prompt + "\nRemember: output ONLY one JSON inside the tags.",
                             num_beams=3, max_new_tokens=512)
            try:
                msg = _extract_json(raw2)
            except Exception:
                # Try 3: repair pass
                try_raw = raw2 if isinstance(raw2, str) else raw
                try:
                    msg = _coerce_to_json(try_raw, tools_text)
                except Exception:
                    try:
                        msg = _coerce_to_json(raw, tools_text)
                    except Exception:
                        msg = {"final":{
                            "summary":"Could not coerce model output into valid JSON after multiple attempts.",
                            "next_actions":"Rerun the step; consider simplifying the GOAL or calling a specific tool directly."
                        }}

        # Validate shape; if invalid, one more repair attempt
        if not _is_valid_shape(msg):
            try:
                msg = _coerce_to_json(json.dumps(msg, ensure_ascii=False), tools_text)
            except Exception:
                msg = {"final":{
                    "summary":"Model returned invalid shape; finalized to prevent crash.",
                    "next_actions":"Retry or call evaluate_pdf/evaluate_inline directly."
                }}

        # Final? Enforce 1 tool
        if "final" in msg:
            if tools_used == 0:
                history.append({"assistant": msg, "observation": {"error": "Final rejected: call at least one tool before final."}})
                continue
            return {"summary": msg["final"], "state": STATE, "history": history}

        # Tool call?
        if "tool_call" in msg:
            name = msg["tool_call"].get("name")
            args = msg["tool_call"].get("args", {})
            spec = TOOL_REGISTRY.get(name)
            if not spec:
                history.append({"assistant": msg, "observation": {"error": f"Unknown tool: {name}"}})
                continue
            try:
                result = spec["fn"](**args)
                tools_used += 1
            except TypeError as e:
                result = {"error": f"Bad args for {name}: {e}"}
            except Exception as e:
                result = {"error": f"Tool {name} failed: {type(e).__name__}: {e}"}
            history.append({"assistant": msg, "observation": result})
            continue

        # Neither final nor tool_call  record error and continue
        history.append({"assistant": msg, "observation": {"error": "Invalid agent message; expected tool_call or final"}})

    # Max steps reached
    return {
        "summary": {"summary": "Max steps reached.", "next_actions": "Consider rerun or call evaluate_pdf/evaluate_inline directly."},
        "state": STATE,
        "history": history,
    }

# Example (single-pair smoke test expected to pass now):
# STATE["qas"].clear(); STATE["results"].clear()
# goal = ("Solve this single problem and grade the student. "
#         "PROBLEM: Compute det([[1,2],[3,4]]). STUDENT_ANSWER: -2. Use tools and then finalize.")
# result = run_agent(goal)
# print(json.dumps(result["summary"], indent=2))
# print("FIRST RESULT:", json.dumps(STATE["results"][:1], indent=2))
```
